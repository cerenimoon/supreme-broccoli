```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('C:/Users/user/Desktop/people.png') #rgb image
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
peoplegray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(peoplegray)
```

```python
cv2.imwrite('graypeople.png',peoplegray) #write the image
```

```python
piece = image[500:1200, 700:1200] #numpy slicing
```

```python
plt.imshow(piece)
```

```python
piece.shape
```

```python
piece2 = image[300:1300, 1800:2500] #numpy sciling
piece2 = cv2.resize(piece2, (500, 700)) #resze the image
```

```python
plt.imshow(piece2) #show the image
```

```python
piece2.shape #image must be same size as counterpart image for logical operations
```

```python
andimage = cv2.bitwise_and(piece, piece2)
```

```python
plt.imshow(andimage)
```

```python
cv2.imwrite('andimage.png', andimage)
```

```python
orimage = cv2.bitwise_or(piece, piece2)
```

```python
plt.imshow(orimage)
```

```python
cv2.imwrite('orimage.png', orimage)
```

```python
xorimage = cv2.bitwise_xor(piece, piece2)
```

```python
plt.imshow(xorimage)
```

```python
cv2.imwrite('xorimage.png', xorimage)
```

```python
histr = cv2.calcHist([peoplegray], [0], None, [256], [0,256]) #cv2 histogram with 256 intensity levels
```

```python
plt.figure() #figure of histogram
plt.title("Grayscale Histogram") #histogram title
plt.xlabel("grayscale value") #histogram x - axis title
plt.ylabel("pixels") #histogram y - axis title
plt.plot(histr) #plot the histogram
plt.show() #show the histogram 
plt.savefig('imghistogram.png')
```
